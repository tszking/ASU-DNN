```python
import pandas as pd
import random 
import numpy as np
from tqdm import tqdm

# filename = 'data/RAW20220307.csv'
filename1 = 'data/data-2104-2108.csv'
filename2 = 'data/data-2109-2111.csv'
filename3 = 'data/data-2112-2202.csv'

df1 = pd.read_csv(filename1)
df2 = pd.read_csv(filename2)
df3 = pd.read_csv(filename3)

df = pd.concat([df1, df2, df3])
```

```python
df = df.rename(columns={'cc': 'cnt'})
```

```python
setODT = set(list([(o, d, date, hour) for o, d, date, hour in zip(df['origin_station_name'], df['destination_station_name'], df['origin_date'], df['hour_'])]))
```

```python
useful_vars = ['origin_station_name', 'destination_station_name', 'origin_date', 'hour_', 'normal_ratio', 'eld_ratio', 'transfer_time', 'avg_time', 'full_price', 'via_stations', 'hourly_cnt', 'od_ratio', 'cnt']
# useful_vars = ['origin_station_name', 'destination_station_name', 'origin_date', 'hour_', 'normal_ratio', 'eld_ratio', 'transfer_time', 'avg_time', 'full_price', 'via_stations', 'hourly_cnt', 'od_ratio', 'cnt']

static_vars = ['hour_', 'normal_ratio', 'eld_ratio']

dict_vars_map1 = {'transfer_time': 'transfer_time1',
                  'avg_time': 'avg_time1',
                  'full_price': 'full_price1', 
                  'via_stations': 'via_stations1', 
                  'hourly_cnt': 'hourly_cnt1',
                  'od_ratio': 'od_ratio1',
                  'cnt': 'cnt1'
                  }

dict_vars_map2 = {'transfer_time': 'transfer_time2', 
                  'avg_time': 'avg_time2', 
                  'full_price': 'full_price2', 
                  'via_stations': 'via_stations2', 
                  'hourly_cnt': 'hourly_cnt2', 
                  'od_ratio': 'od_ratio2',
                  'cnt': 'cnt2'
                  }

dict_vars_map3 = {'transfer_time': 'transfer_time3',  
                  'avg_time': 'avg_time3', 
                  'full_price': 'full_price3', 
                  'via_stations': 'via_stations3', 
                  'hourly_cnt': 'hourly_cnt3', 
                  'od_ratio': 'od_ratio3',
                'cnt': 'cnt3'
                  }

dict_vars_map4 = {'transfer_time': 'transfer_time4',  
                  'avg_time': 'avg_time4', 
                  'full_price': 'full_price4', 
                  'via_stations': 'via_stations4', 
                  'hourly_cnt': 'hourly_cnt4', 
                  'od_ratio': 'od_ratio4',
                'cnt': 'cnt4'
                  }

dict_vars_map5 = {'transfer_time': 'transfer_time5',  
                  'avg_time': 'avg_time5', 
                  'full_price': 'full_price5', 
                  'via_stations': 'via_stations5', 
                  'hourly_cnt': 'hourly_cnt5', 
                  'od_ratio': 'od_ratio5',
                'cnt': 'cnt5'
                  }

listDictVarsMap = [dict_vars_map1, dict_vars_map2, dict_vars_map3, dict_vars_map4, dict_vars_map5]
```

```python
len(setODT)
```

```python
listDfAll = []

# count = 0
# for odt in setODT:
for odt in tqdm(setODT):
    o, d, date, hour = odt
    dfTemp = df[(df['origin_station_name']==o) & (df['destination_station_name']==d) & (df['origin_date']==date) & (df['hour_']==hour)]
    n = dfTemp.shape[0]
    rnd_column_idx = random.sample(range(5), n)
    if n > 1:
        n_idx_shuffle = np.arange(1,n+1)
        np.random.shuffle(n_idx_shuffle)

        listDfTemp = [dfTemp[dfTemp['cnt_rnk']==n_idx_shuffle[i]][useful_vars].rename(columns=listDictVarsMap[rnd_column_idx[i]]) for i in range(n)]

        # 生成宽表
        dfAll = listDfTemp[0]
        for i in range(1, n):
            dfAll = pd.merge(dfAll,listDfTemp[i], on=['origin_station_name', 'destination_station_name', 'origin_date', 'hour_', 'normal_ratio', 'eld_ratio'])
        listDfAll.append(dfAll)
        # count += 1
        # print(count)

dfDatasets = pd.concat(listDfAll, axis=0) #.to_csv('check.csv')
```

```python
dfDatasets
```

```python
dfDatasets['cnt1'] = dfDatasets['cnt1'].fillna(0)
dfDatasets['cnt2'] = dfDatasets['cnt2'].fillna(0)
dfDatasets['cnt3'] = dfDatasets['cnt3'].fillna(0)
dfDatasets['cnt4'] = dfDatasets['cnt4'].fillna(0)
dfDatasets['cnt5'] = dfDatasets['cnt5'].fillna(0)

dfDatasets['total_cnt'] = dfDatasets['cnt1'] + dfDatasets['cnt2'] + dfDatasets['cnt3'] + dfDatasets['cnt4'] + dfDatasets['cnt5']
```

```python
# dfTruth = dfDatasets[['origin_station_name', 'destination_station_name', 'origin_date', 'hour_', 'od_ratio1', 'od_ratio2', 'od_ratio3', 'od_ratio4', 'od_ratio5']]
dfTruth = dfDatasets[['origin_station_name', 'destination_station_name', 'origin_date', 'hour_', 'cnt1', 'cnt2', 'cnt3', 'cnt4', 'cnt5']]
dfTruth = dfTruth.fillna(0)

dfTrain = dfDatasets.drop(columns=['origin_station_name', 'destination_station_name', 'origin_date', 
                                    'od_ratio1', 'od_ratio2', 'od_ratio3', 'od_ratio4', 'od_ratio5',
                                    'cnt1', 'cnt2', 'cnt3', 'cnt4', 'cnt5'
                                    ])
# dfTrain = dfTrain.fillna(-1)
dfTrain['transfer_time1'] = dfTrain['transfer_time1'].fillna(99)
dfTrain['transfer_time2'] = dfTrain['transfer_time2'].fillna(99)
dfTrain['transfer_time3'] = dfTrain['transfer_time3'].fillna(99)
dfTrain['transfer_time4'] = dfTrain['transfer_time4'].fillna(99)
dfTrain['transfer_time5'] = dfTrain['transfer_time5'].fillna(99)

dfTrain['avg_time1'] = dfTrain['avg_time1'].fillna(99)
dfTrain['avg_time2'] = dfTrain['avg_time2'].fillna(99)
dfTrain['avg_time3'] = dfTrain['avg_time3'].fillna(99)
dfTrain['avg_time4'] = dfTrain['avg_time4'].fillna(99)
dfTrain['avg_time5'] = dfTrain['avg_time5'].fillna(99)

dfTrain['full_price1'] = dfTrain['full_price1'].fillna(99)
dfTrain['full_price2'] = dfTrain['full_price2'].fillna(99)
dfTrain['full_price3'] = dfTrain['full_price3'].fillna(99)
dfTrain['full_price4'] = dfTrain['full_price4'].fillna(99)
dfTrain['full_price5'] = dfTrain['full_price5'].fillna(99)

dfTrain['via_stations1'] = dfTrain['via_stations1'].fillna(99)
dfTrain['via_stations2'] = dfTrain['via_stations2'].fillna(99)
dfTrain['via_stations3'] = dfTrain['via_stations3'].fillna(99)
dfTrain['via_stations4'] = dfTrain['via_stations4'].fillna(99)
dfTrain['via_stations5'] = dfTrain['via_stations5'].fillna(99)

dfTrain['hourly_cnt1'] = dfTrain['hourly_cnt1'].fillna(0)
dfTrain['hourly_cnt2'] = dfTrain['hourly_cnt2'].fillna(0)
dfTrain['hourly_cnt3'] = dfTrain['hourly_cnt3'].fillna(0)
dfTrain['hourly_cnt4'] = dfTrain['hourly_cnt4'].fillna(0)
dfTrain['hourly_cnt5'] = dfTrain['hourly_cnt5'].fillna(0)

listTrainColumns = ['hour_', 'normal_ratio', 'eld_ratio', 'total_cnt',
                    'transfer_time1', 'avg_time1', 'full_price1', 'via_stations1', 'hourly_cnt1', 
                    'transfer_time2', 'avg_time2', 'full_price2', 'via_stations2', 'hourly_cnt2',
                    'transfer_time3', 'avg_time3', 'full_price3', 'via_stations3', 'hourly_cnt3',
                    'transfer_time4', 'avg_time4', 'full_price4', 'via_stations4', 'hourly_cnt4',
                    'transfer_time5', 'avg_time5', 'full_price5', 'via_stations5', 'hourly_cnt5'
                    ]
dfTrain = dfTrain[listTrainColumns]

dfTrain.to_csv('train_set.csv')
```
